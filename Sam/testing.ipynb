```python
import pandas as pd
import numpy as np
from sklearn import svm

import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)


results = pd.read_csv('../data/RegularSeasonDetailedResults.csv')
```

```python
results['Wfga'] >results['Wfgm'].values.any()
```

```python
#offensive rebounds ('wor') and field goals made ('wfgm')
off_reb = results['Wor']
fgm = results['Wfga']
plt.xlabel('off_reb')
plt.ylabel('wfga')
plt.scatter(off_reb, fgm)

plt.show()
print(off_reb.corr(fgm))
```

```python
#blocks and defensive rebounds
def_reb = results['Wdr']
blocks = results['Wblk']
plt.xlabel('def_reb')
plt.ylabel('blocks')
plt.scatter(def_reb,blocks)

plt.show()
print(def_reb.corr(blocks))
```
